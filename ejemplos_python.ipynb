```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,50)
y=np.linspace(0,5,50)


X,Y = np.meshgrid(x,y)
Z = np.cos(X) + np.sin(Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


plt.show()
```

```python
#defino nuevas variables
x=np.linspace(0,2*np.pi,30)
```

```python
fig,axes=plt.subplots(2,2,figsize=(20,20))
axes[0,1].scatter(x,np.sin(x),edgecolors='k',c='k')
axes[0,1].plot(x,np.sin(x),c='k')
axes[0,1].set_title('Seno')
axes[1,1].plot(x,np.cos(x))
axes[1,1].set_title('Cos')
axes[0,0].set_ylabel('y')
axes[0,0].set_xlabel('x')
axes[1,0].set_ylabel('y')
axes[1,0].set_xlabel('x')
plt.tight_layout()
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.grid()
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
#practico: grafico de dispersion e^3x arriba
#grafico dispersion + lineas cos 2y
x=np.linspace(0,5,100)
fig,axes=plt.subplots(2,1,figsize=(10,10))
axes[0].scatter(x,np.exp(3*x),s=8,edgecolor='k',label='$e^{3x}$')
axes[0].set_ylabel("y")
axes[0].set_xlabel("x")
axes[0].legend()
axes[1].scatter(x,np.cos(2*x),s=8,edgecolor='k',label='${cos(2x)}$')
#axes[1].legend()
#tambien:
axes[1].legend(bbox_to_anchor=(0.85, 1), loc='upper left') #bbox_to_anchor mov x,y respect.
axes[1].plot(x,np.cos(2*x))
axes[1].errorbar(x,np.cos(2*x),yerr=dy,c='blue')
axes[1].set_ylabel("y")
axes[1].set_xlabel("x")
```
